```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
dataset1 = pd.read_csv('dataset1.csv')
dataset2 = pd.read_csv('dataset2.csv')
dataset3 = pd.read_csv('dataset3.csv')
```

```python
def descriptive_analysis_1(data):
    """
    Investigation 1: Calculate basic descriptive statistics for numerical columns
    """
    print("Descriptive Analysis 1: Basic Statistics")
    print(data.describe())
    
    # Visualize distribution of a key variable
    key_variable = data.select_dtypes(include=[np.number]).columns[1]  
    plt.figure(figsize=(10, 6))
    data[key_variable].hist()
    plt.title(f'Distribution of {key_variable}')
    plt.xlabel(key_variable)
    plt.ylabel('Frequency')
    plt.show()
```

```python
def descriptive_analysis_2(data):
    """
    Investigation 2: Analyze correlation between two numerical variables
    """
    print("Descriptive Analysis 2: Correlation Analysis")
    numeric_cols = data.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) < 3:
        print("Not enough numeric columns for correlation analysis")
        return
    
    col1, col2 = numeric_cols[1],numeric_cols[2]  # Choose first two numeric columns
    correlation = data[col1].corr(data[col2])
    print(f"Correlation between {col1} and {col2}: {correlation}")
    
    plt.figure(figsize=(10, 6))
    plt.scatter(data[col1], data[col2])
    plt.title(f'Scatter plot: {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

```python
def inferential_analysis_1(data):
    """
    Investigation 3: Perform a t-test to compare means of two groups
    """
    print("Inferential Analysis 1: Two-Sample T-Test")
    numeric_cols = data.select_dtypes(include=[np.number]).columns
    categoric_cols = data.select_dtypes(include=['object']).columns
    
    if len(numeric_cols) == 0 or len(categoric_cols) == 0:
        print("Not enough suitable columns for t-test")
        return
    
    numeric_col = numeric_cols[0]  # Choose first numeric column
    categoric_col = categoric_cols[0]  # Choose first categorical column
    
    groups = data[categoric_col].unique()
    if len(groups) < 2:
        print("Not enough groups for comparison")
        return
    
    group1 = data[data[categoric_col] == groups[0]][numeric_col]
    group2 = data[data[categoric_col] == groups[1]][numeric_col]
    
    t_stat, p_value = stats.ttest_ind(group1, group2)
    print(f"T-statistic: {t_stat}, p-value: {p_value}")
```

```python
def inferential_analysis_2(data):
    """
    Investigation 4: Perform a chi-square test of independence
    """
    print("Inferential Analysis 2: Chi-Square Test of Independence")
    categoric_cols = data.select_dtypes(include=['object']).columns
    
    if len(categoric_cols) < 2:
        print("Not enough categorical columns for chi-square test")
        return
    
    col1, col2 = categoric_cols[:2]  # Choose first two categorical columns
    contingency_table = pd.crosstab(data[col1], data[col2])
    
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
    print(f"Chi-square statistic: {chi2}, p-value: {p_value}")
```

```python
for i, dataset in enumerate([dataset1, dataset2, dataset3], 1):
    print(f"\nAnalyzing Dataset {i}")
    descriptive_analysis_1(dataset)
    descriptive_analysis_2(dataset)
    inferential_analysis_1(dataset)
    inferential_analysis_2(dataset)
```
