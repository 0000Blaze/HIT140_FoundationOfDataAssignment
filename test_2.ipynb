```python
descriptive_stats = dataset3[['Optm', 'Usef', 'Relx']].describe()
print(descriptive_stats)
```

```python
# Correlation between 'Optm' and 'Relx'
correlation_optm_relx = dataset3['Optm'].corr(dataset3['Relx'])
print("Correlation between Optimism and Relaxation:", correlation_optm_relx)
```

```python
from scipy import stats

# Performing a t-test between 'Optm' and 'Relx'
t_statistic, p_value = stats.ttest_ind(dataset3['Optm'], dataset3['Relx'])

print(f"T-statistic: {t_statistic}, P-value: {p_value}")
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Optm', y='Usef', data=dataset3, estimator=np.mean)
plt.title('Average Sales by Optm')
plt.ylabel('Average Sales')
plt.xlabel('Usef')
plt.show()
```
